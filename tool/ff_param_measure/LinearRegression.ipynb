```python
#%matplotlib inline
import numpy as np
import pandas as pd
#dtype_={"V_L":np.float, "V_R":np.float, "velocity":np.float, "acc":np.float, "omega":np.float,"alpha":np.float}
run_data = pd.read_csv("search_data.csv", sep=",",dtype=float)
#run_data = run_data.astype({'acc': float})
#print(run_data)
#print(run_data.dtypes)
run_data.plot( y=['V_L', 'V_R'], figsize=(16,4), alpha=0.5)
run_data.plot( y=['velocity'], figsize=(16,4), alpha=0.5)
run_data.plot( y=['acc'], figsize=(16,4), alpha=0.5)
run_data.plot( y=['omega'], figsize=(16,4), alpha=0.5)
run_data.plot( y=['alpha'], figsize=(16,4), alpha=0.5)
```

```python
from sklearn import linear_model
clf = linear_model.LinearRegression()
 
#  "density ()" 
X = run_data.drop(run_data.columns[[0, 1,2,3]], axis=1)
 
#  "alcohol ()" 
Y = run_data["V_L"].as_matrix()
 
# 
clf.fit(X, Y)
 
# 
print(clf.coef_)
 
#  ()
print(clf.intercept_)
 
# 
print(clf.score(X, Y))
```
